```python
import tensorflow
```

```python
df=pd.read_csv('C:\\Users\\NITJ\Downloads\\mnist_train.csv\\mnist_train.csv', delimiter=',', dtype=int)
df
```

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
# Normalize
X=X/255
```

```python
y
```

```python
X.shape
```

```python
X_train=X
y_train=y
```

```python
pd.set_option("display.max_columns",20)
pd.set_option("display.max_rows",10)
```

```python
pd.DataFrame(X_train)
```

```python
# Part 2 - Now let's make the ANN!
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Create a new Sequential object
model = Sequential()

# Create the input layer and 1st hidden layer with 400 nodes
model.add(Dense(400,kernel_initializer ='uniform', activation = 'relu', input_dim = 784))

# Create another hidden layer with 100 nodes
model.add(Dense(100, activation="relu"))
          

# Create an output layer, 10 nodes
model.add(Dense(10, activation="softmax"))

# Compile model
model.compile(optimizer = 'adam' , loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
from keras.utils import np_utils
dummy_y=np_utils.to_categorical(y_train)
dummy_y
```

```python
model.fit(X_train,dummy_y,batch_size = 10 ,nb_epoch=5)
```

```python
train_accuracy = model.evaluate(X_train, dummy_y)
print(train_accuracy[1])
```
